```python
import sys, os
# sys.stdout = open(os.devnull, 'w')
from wikifier import utils
from datamart.entries_new import D3MDatamart, D3MJoinSpec
import pandas as pd
d3mDatamart = D3MDatamart()
# this is our original input dataset
inputs_ds_loc = "/Users/minazuki/Desktop/studies/master/2018Summer/data/customize/\
DA_poverty_estimation/TRAIN/dataset_TRAIN/datasetDoc.json"
# pd.set_option('display.max_columns', None)
original_dataset = utils.load_d3m_dataset(inputs_ds_loc)
```

```python
original_dataset['learningData'].head()
```

```python
# this is our input training dataset, after sending to wikifier
wikifier_dataset_use_all_columns = utils.wikifier_for_d3m_all(input_ds=original_dataset).value
utils.pretty_print(wikifier_dataset_use_all_columns,"wikifier")
```

```python
# wikifier_dataset = utils.wikifier_for_d3m_fips(input_ds=original_dataset).value
search_result = d3mDatamart.search_with_data(supplied_data=wikifier_dataset_use_all_columns)
```

```python
# wiki_search_result.display()
utils.print_search_results(search_result)
```

```python
wikifier_dataset_use_all_columns['learningData']
```

```python
wiki_search_result = search_result[0]
augmented_wikifier_dataset = wiki_search_result.augment(supplied_data=wikifier_dataset_use_all_columns)
# we get the returnede augmented dataset
augmented_wikifier_dataset.keys()
# here is the detail dataframe inside our augmented dataset with wikidata
utils.pretty_print(augmented_wikifier_dataset,"wiki_augment")
```

```python
# we search a second time base on this augmented dataset
search_result2 = d3mDatamart.search_with_data(supplied_data=augmented_wikifier_dataset)
# here we display the search results in a human vision
utils.print_search_results(search_result2)
```

```python
datamart_search_result = search_result2[6]
# run the download function to get the dataset and corresponding joining pairs hint
augmented_datamart_dataset_download = datamart_search_result.download(supplied_data=augmented_wikifier_dataset)
utils.pretty_print(augmented_datamart_dataset_download,"download")
```

```python
augmented_datamart_dataset_augment = datamart_search_result.augment(supplied_data=augmented_wikifier_dataset)
```
